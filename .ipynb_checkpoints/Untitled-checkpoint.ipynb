```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patch
import math
```

```python
fig, ax = plt.subplots()
plt.xlim(-1, 1)
plt.ylim(-1, 1)
ax.axis('equal')
#ax.axis('off')

square = patch.Rectangle((-1, -1), 2, 2, color='black', fill=False)
ax.add_patch(square)

circle = plt.Circle((0, 0), 1, color='black', fill=False, clip_on=False)
ax.add_artist(circle)

inside_circle = 0
outside_circle = 0
for i in range(100):
    
    # random coordinate
    x_rand = (np.random.random() - 0.5)*2 
    y_rand = (np.random.random() - 0.5)*2
    
    if math.sqrt((x_rand - 0)**2 + (y_rand - 0)**2) < 1:
        ax.plot(x_rand, y_rand, 'rx')
        inside_circle+=1
    else:
        ax.plot(x_rand, y_rand, 'gx')
        outside_circle+=1

pi_aprox = (inside_circle/(inside_circle+outside_circle))*4

print(pi_aprox)
        
    
    
```

```python
real = np.linspace(-2, 2, 1000)
imag = np.linspace(-1, 1, 1000)
x_axis = []
y_axis = []
max_iter = 10

def mandel(c, max_iter):
    '''
    Iterates max_iter times over the F(Z) function.
    '''
    z = 0
    for i in range(0, max_iter):
        z = z**2 + c

    return z

def mandel_set(x, y, max_iter):
    '''
    The distance to 0 must be <2 for any point in the Mandelbrot set. The
    distance is the absolute alue or modulus of complex number c.
        c = a +bi
    Where a is the real part (x-axis) and b is the imaginary part (y-axis)
    '''
    for i in x:
        for z in y:
            a = complex(i+z*((-1)**(-1/2)))
            p
            m = mandel(a, max_iter)
            if abs(m) < 2:
                x_axis.append(i)
                y_axis.append(z)

mandel_set(real, imag, max_iter)

plt.scatter(x_axis, y_axis)
plt.show()
```

```python
import random

T = 100000
N = 100

def randomcomplex():
    a, b = np.random.uniform(-2, 2, 2)
    c = complex(a,b)
    
    return c, a, b

def f_iter(c, N):
    #print(c)
    
    z=0
    for i in range(0, N):
        z=z*z + c
        #print(z)
    return z

fig, ax = plt.subplots()
plt.xlim(-2.1, 1)
plt.ylim(-1.3, 1.3)

in_mandelbrot=0
outside_mandelbrot=0
for i in range(T):
    c, a, b = randomcomplex()
    z = f_iter(c, N)
    if abs(z) <= 2:
        ax.plot(a, b, 'gx')
        in_mandelbrot+=1
    else:
        #ax.plot(a, b, 'rx')
        outside_mandelbrot+=1
    #print(z)

mandel_area = (in_mandelbrot/(in_mandelbrot+outside_mandelbrot))*16
print(mandel_area)
    
```
